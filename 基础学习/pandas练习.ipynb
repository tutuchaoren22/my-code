```python
df1 = pd.DataFrame(np.ones((3,4))*0, columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1, columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4))*2, columns=['a','b','c','d'])
res = pd.concat([df1, df2, df3], axis=0) #axis = 0为纵向合并 =1 为横向合并
print(res)
res = pd.concat([df1,df2,df3],axis = 0, ignore_index = True) #将index变成0-8
print(res)
df1 = pd.DataFrame(np.ones((3,4))*0, columns=['a','b','c','d'], index=[1,2,3])
df2 = pd.DataFrame(np.ones((3,4))*1, columns=['b','c','d','e'], index=[2,3,4])
res = pd.concat([df1, df2], axis=0, join='outer') #join='outer'为预设值，因此未设定任何参数时，函数默认join='outer'。此方式是依照column来做纵向合并，有相同的column上下合并在一起，其他独自的column个自成列，原本没有值的位置皆以NaN填充
print(res)
res = pd.concat([df1, df2], axis=0, join='inner') #只有相同的column合并在一起，其他的会被抛弃
print(res)
res = pd.concat([df1, df2], axis=1, join_axes=[df1.index])
print(res)
df1 = pd.DataFrame(np.ones((3,4))*0, columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1, columns=['a','b','c','d'])
res = df1.append(df2, ignore_index=True) #append只有纵向合并
print(res)
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                             'A': ['A0', 'A1', 'A2', 'A3'],
                             'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                              'C': ['C0', 'C1', 'C2', 'C3'],
                              'D': ['D0', 'D1', 'D2', 'D3']})
print(left)
print(right)
res = pd.merge(left,right, on = 'key')
print(res)
```

```python
#合并时有四种方法 how = ['left','right','inner','outer']
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                      'key2': ['K0', 'K1', 'K0', 'K1'],
                      'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                       'key2': ['K0', 'K0', 'K0', 'K0'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
print(left)
print(right)
res = pd.merge(left,right, on = ['key1','key2'],how = 'inner') #相同的连接
print(res)
res = pd.merge(left,right, on = ['key1','key2'],how = 'outer') #所有的都连接 缺失的补nan
print(res)
res = pd.merge(left,right, on = ['key1','key2'],how = 'left') #以前面的连接
print(res)
res = pd.merge(left,right, on = ['key1','key2'],how = 'right') #以后面的连接
print(res)
```

```python
df1 = pd.DataFrame({'col1':[0,1], 'col_left':['a','b']})
df2 = pd.DataFrame({'col1':[1,2,2],'col_right':[2,2,2]})
print(df1)
print(df2)
res = pd.merge(df1,df2, on = 'col1', how = 'outer', indicator = True) #indicator 将合并的记录放在新的一列
print(res)
res = pd.merge(df1,df2, on = 'col1', how = 'outer', indicator = 'a') #可以定义这列的名称
print(res)
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2'])
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                     index=['K0', 'K2', 'K3'])
print(left)
print(right)
res = pd.merge(left,right,left_index = True,right_index = True,how = 'inner') #以index进行合并
print(res)
res = pd.merge(left,right,left_index = True,right_index = True,how = 'outer')
```
